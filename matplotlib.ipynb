```python
x1= ['Jan','Feb','March','April','May','June']
y1= [120,34,56,110,50,90]
x2= ['Jan','Feb','March','April','May','June']
y2= [60,130,90,45,150,90]
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.plot(x1,y1 ,color='r',marker='o', label = 'Y 2021')
plt.plot(x1,y2 ,color='b',marker='*', label = 'Y 2020')
plt.legend()

plt.title('Month Vs Rainfall')
plt.show()
```
