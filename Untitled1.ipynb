```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = [23,46,34,78,101,92,56]

mean= np.mean(data)
median= np.median(data)
skewness=stats.skew(data)
std_dev= np.std(data)

fig,(ax1,ax2) = plt.subplots(1,2,figsize=(12,5))

#plot histogram of data

ax1.hist(data, bins=range(20,90,10), edgecolor='black', alpha=0.7,color='skyblue')
ax1.axvline(mean, color='red', linestyle='--', label=f'Mean:{mean:.2f}')
ax1.axvline(median, color='green', linestyle='--', label=f'Median:{median}')
ax1.set_title('Histogram of Data')
ax1.set_xlabel('Value')
ax1.set_ylabel('Frequency')
ax1.legend()
ax1.grid(axis='y', alpha=0.4)

#plot boxplot

plt.subplot(1,2,2)
plt.boxplot(data, vert=True, patch_artist=True, boxprops=dict(facecolor='lightcoral'))
plt.title('BoxPlot of Data')
plt.ylabel('Value')
plt.grid(axis='y', alpha=0.4)


plt.tight_layout()
plt.show()

print("Statistical Summary:")
print(f"Sample Size: {len(data)}")
print(f"Mean: {mean:.2f}")
print(f"Median: {median}")
print(f"Skewness: {skewness:.3f}")
print(f"Standard Deviation: {std_dev:.2f}")
```

```python

total_cards = 52
cards_to_draw = 4
total_kings = 4
kings_to_draw = 4
.
total_outcomes = math.comb(total_cards, cards_to_draw)

favorable_outcomes = math.comb(total_kings, kings_to_draw)


probability = favorable_outcomes / total_outcomes


print(f"Total possible 4-card hands: {total_outcomes:,}")
print(f"Number of ways to draw 4 kings: {favorable_outcomes}")
print(f"Probability of drawing 4 kings: 1 in {int(total_outcomes / favorable_outcomes):,}")
print(f"As a decimal: {probability}")
print(f"As a percentage: {probability:.6%}")
```
